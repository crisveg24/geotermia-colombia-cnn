Se crea una imagen del nevado del ruiz

```python
import ee
import geemap

# Inicializa Earth Engine
ee.Initialize()

# Carga la imagen de emisividad ASTER
aster = ee.Image("NASA/ASTER_GED/AG100_003")

# Define una región de interés (cerca del volcán Galeras, por ejemplo)
roi = ee.Geometry.Point([-77.36, 1.22]).buffer(5000)

# Crea el mapa
Map = geemap.Map()
Map.centerObject(roi, 10)
Map.addLayer(aster.select('emissivity_band13'), {'min': 0.8, 'max': 1.0}, 'Emisividad B13')
Map
```

codigo para guardar tres imagenes etiquedas como potnencial geotemico en colombia

```python
import os
import geemap

# Crear directorio de descarga
download_dir = '/mnt/data/geotermia_imagenes'
os.makedirs(download_dir, exist_ok=True)

# Zonas geotérmicas de Colombia (coordenadas aproximadas)
zones = {
    "Nevado_del_Ruiz": ee.Geometry.Point([-75.3222, 4.8951]),
    "Volcan_Purace": ee.Geometry.Point([-76.4036, 2.3206]),
    "Paipa_Iza": ee.Geometry.Point([-73.1124, 5.7781])
}

# Cargar imagen ASTER de emisividad
dataset = ee.Image('NASA/ASTER_GED/AG100_003')
band = dataset.select('emissivity_band10')  # Banda 10 de emisividad

# Función para descargar cada imagen
def download_image(name, geom):
    roi = geom.buffer(5000)  # Buffer de 5 km alrededor del punto
    image = band.clip(roi)

    path = os.path.join(download_dir, f"{name}.tif")
    geemap.ee_export_image(
    band.clip(roi),   # directamente la imagen
    filename=path,
    scale=100,
    region=roi,
    file_per_band=False
    )

    print(f"✅ Imagen descargada: {path}")

# Descargar imágenes
for name, geom in zones.items():
    download_image(name, geom)
```

Codigo para visualziar imagenes con rasterio y matplotlib

```python
import rasterio
import matplotlib.pyplot as plt

# Cambia esta ruta por la ubicación de tu archivo .tif descargado
ruta_imagen = '/mnt/data/geotermia_imagenes/Nevado_del_Ruiz.tif'

# Abrir la imagen con rasterio
with rasterio.open(ruta_imagen) as src:
    # Leer la primera banda (por ejemplo)
    banda = src.read(1)

    # Mostrar la imagen
    plt.figure(figsize=(8, 6))
    plt.imshow(banda, cmap='inferno')  # puedes usar otros colormaps como 'gray', 'viridis'
    plt.colorbar(label='Valor de emisividad')
    plt.title('Imagen ASTER Emisividad - Nevado del Ruiz')
    plt.axis('off')  # quitar ejes si prefieres
    plt.show()
```
